```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
gps_df = pd.read_csv("GPS_measurement.csv")
```

```python
gps_df.head()
```

```python
processed_gps_df = pd.DataFrame()
processed_gps_df['datetime'] = gps_df['timestamp'].map(lambda x : datetime.fromtimestamp(x // 1000000000))
processed_gps_df['lat'] = gps_df['lat']/1e7
processed_gps_df['lon'] = gps_df['lon']/1e7
processed_gps_df['alt'] = gps_df['alt']/1000
processed_gps_df['vel'] = gps_df['vel']/100
```

```python
processed_gps_df.head()
```

```python
plt.plot(processed_gps_df['lat'], processed_gps_df['lon'])
plt.show()
```
